```python
import pandas as pd
```

```python
df_pop = pd.read_csv('/content/population_total.csv')
```

Making Pivot table - pivot() - no aggregation only reshape

```python
df_pop.dropna(inplace=True)
```

```python
df_new = df_pop.pivot(index = 'year',columns = 'country',values = 'population')
```

```python
df_new
```

```python
df_pivot = df_new[['China','India','United States','Indonesia','Brazil']]
```

Line Plot

```python
df_pivot.plot(kind = 'line', xlabel = 'Year', ylabel = 'Population',title = 'Population by Year for Countries')
```

```python
df_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_2020
```

```python
df_2020 = df_2020.T
```

```python
df_2020.plot(kind='bar',xlabel='Country',ylabel ='Population',color = 'orange',title = 'Population in 2020',figsize = (8,4))
```

```python
df_pivot
```

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
df_pivot_sample.plot(kind = 'bar',xlabel='Country',ylabel ='Population',color = 'orange',figsize = (8,4))
```
