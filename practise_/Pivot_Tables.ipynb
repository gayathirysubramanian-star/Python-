pivot_table()- Spreadsheets in Excel

```python
df_sales = pd.read_excel('/content/supermarket_sales.xlsx')
df_sales
```

```python
df_sales.drop(columns = ['Date','Time'],inplace = True)
```

```python
df_sales.columns
```

```python
df_sales.pivot_table(index='Gender',values=df_sales.select_dtypes('number').columns, aggfunc='sum')
```

```python
 df_sales.pivot_table(index='Gender',values=['Quantity','Total'], aggfunc='sum')
```

```python
 df_sales.pivot_table(index='Gender',columns ='Product line',values=['Quantity','Total'], aggfunc='sum')
```
